```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime as dt
```

```python
class Stats_Ticker_Of_Choice_YTD:

    def __init__(self, ticker, start):
        self.ticker = ticker
        self.start = start


        # Call Methods
        self.StatsMaker()


    def StatsMaker(self):
        # Download stock data
        self.df = yf.download(self.ticker, self.start, dt.date.today())

        # Calculate percentage change
        self.df['pct_change'] = self.df['Close'].pct_change()

        # Plot KDE
        sns.kdeplot(self.df['pct_change'].dropna())
        plt.title(f'Normalized distibution from {self.start} till today of {self.ticker}.')
        plt.xlabel('The > Timeframe, the more normalized destribution')
        plt.ylabel('Density')
        plt.show()
```

```python
# CHOOSE YOUR TICKER SND START DATE BELOW

# THE CLOSER THE START DATE TO TODAY --> THE MORE SKEW AND TAILS

ticker = "NVDA"       # Change for your own ticker, accoring to Yahoo finance tickers, format 'SPY'
start = "2024-1-1"     # Change for your own date of choice, format 'yyyy-m-d'

your_ticker_object = Stats_Ticker_Of_Choice_YTD(ticker,start)
```
