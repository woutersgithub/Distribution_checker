```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime as dt
```

```python
class Stats_Ticker_Of_Choice_YTD:

    def __init__(self, ticker, start, end):
        self.ticker = ticker
        self.start = start
        self.end = end


        # Call Methods
        self.StatsMaker()


    def StatsMaker(self):
        # Download stock data
        self.df = yf.download(self.ticker, self.start, self.end)

        # Calculate percentage change
        self.df['pct_change'] = self.df['Close'].pct_change()

        # Plot KDE
        sns.kdeplot(self.df['pct_change'].dropna())
        plt.title(f'Normalized distibution from {self.start} --> {self.end} of {self.ticker}.')
        plt.xlabel('The > Timeframe, the more normalized destribution')
        plt.ylabel('Density')
        plt.show()
```
